```python
from bs4 import BeautifulSoup
import urllib.request
import urllib.parse
import pandas as pd
from pandas import DataFrame
import numpy as np
```

```python
api = 'https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date='
movie_and_score = {}
movielist = []
scorelist = []
datelist = []
for date in pd.date_range('2019-09-10', '2019-09-19', freq='D'):
    url = api+date.strftime('%Y%m%d')
    data = urllib.request.urlopen(url).read()
    soup = BeautifulSoup(data,'html.parser')

#     movielist = soup.select('tbody > tr > td.title > div > a')
#     scorelist = soup.select('tbody > tr> td.point')
#     for movie, score in zip(movielist,scorelist):
#         if movie.get_text() in movie_and_score:
#             list = movie_and_score[movie.get_text()]
#             list.append(score.get_text())
# #             movie_and_score[movie.get_text()]=list    //list의 주소값을 가지고 오기 때문에 바로 딕셔너리 리스트 안에 들어감
#         else:
#             movie_and_score[movie.get_text()] = [score.get_text()]

    movie = [ name.string for name in soup.select('tbody > tr > td.title > div > a')]
    movielist.extend(movie)
    
    score = [ point.string for point in soup.select('tbody > tr> td.point')]
    scorelist.extend(score)
    
    day = [ date for i in range(len(movie))]
    datelist.extend(day)
```

```python
movie_data = {
    '영화': movielist,
    '평점': scorelist,
    '날짜': datelist
}
movie_table = pd.DataFrame(movie_data, columns=['영화','평점','날짜'])
movie_table['평점'] = movie_table['평점'].map(lambda x : float(x))
movie_table['평점'][0]
```

```python
movie_date_score = pd.pivot_table( movie_table, 
                        columns=['영화'],
                        index=['날짜'], 
                        values=['평점'],
                       fill_value='')
movie_date_score.head()
```
